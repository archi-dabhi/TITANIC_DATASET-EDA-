```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Titanic_data.csv')
print(data.head())
```

```python
# Use .describe(), .info(), .value_counts()

print("Summary statistics for all numerical columns:\n")
print(data.describe())

print("\nDataframe information:\n")
data.info()

print("\nSurvival count:\n")
print(data['Survived'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols = ["Survived", "Pclass", "Age", "SibSp", "Parch", "Fare"]

# Pairplot

sns.pairplot(data[numeric_cols], hue="Survived", palette="viridis")
plt.suptitle("Pairplot of Titanic Numerical Features", y=1.02)
plt.show()
print("Observation: Survivors are more common in 1st class and at higher fares. Age shows no strong separation.\n")

```

```python
# Heatmap of correlations

plt.figure(figsize=(8,6))
sns.heatmap(data[numeric_cols].corr(), annot=True, cmap="viridis")
plt.title("Correlation Heatmap")
plt.show()
print("Observation: Pclass is negatively correlated with survival (-0.34), Fare is positively correlated (0.26). Age has weak correlation.")

```

```python
# Histograms

plt.figure(figsize=(8, 5))
sns.histplot(data['Age'].dropna(), bins=30, kde=True, color=sns.color_palette("crest")[2] )
plt.title("Age Distribution of Passengers")
plt.show()
print("Observation: Most passengers are between 20–40 years old, with fewer children and elderly.\n")


# Boxplot - Fare by Passenger Class

plt.figure(figsize=(8, 5))
sns.boxplot(x="Pclass", y="Fare", hue="Pclass", data=data, palette="Set2" )
plt.title("Fare Distribution by Passenger Class")
plt.show()
print("Observation: 1st class fares are much higher than 2nd and 3rd class, with 3rd class fares concentrated at lower values.\n")


# Scatterplot - Age vs Fare by Survival

plt.figure(figsize=(8, 5))
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=data, palette="viridis")
plt.title("Age vs Fare by Survival")
plt.show()
print("Observation: Survivors often paid higher fares. Children have relatively higher survival chances regardless of fare.")
```

```python

# Barplot - Survival Rate by Gender

plt.figure(figsize=(6, 4))
sns.barplot(x="Sex", y="Survived",hue="Sex", data=data, palette="viridis")
plt.title("Survival Rate by Gender")
plt.show()
print("Observation: Females had a much higher survival rate than males (~74% vs ~19%).")
```

```python
print("\n--- Summary of Findings ---")
print("""
- Survival strongly depends on socioeconomic status (Pclass) and gender.
- Fare is positively correlated with survival, indicating better cabins had better access to lifeboats.
- Age has little correlation, but children fared slightly better.
- Cherbourg passengers had the highest survival rate, Southampton the lowest.
- Women and higher-class passengers were prioritized during evacuation.
""")
```
